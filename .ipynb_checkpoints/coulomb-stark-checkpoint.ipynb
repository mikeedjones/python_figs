```python
from scipy.special import ellipj, ellipk

def xi_minus(E,F,Z_1,Z_c):
    return (abs(E)/F)*(np.sqrt(1+Z_1/Z_c)-np.sign(E))

def phi(E,Z_1,Z_c):
    return np.sqrt(abs(E)/2)*(1+Z_1/Z_c)**.25

def m_xi(E,Z_1,Z_c):
    return .5*(1+np.sign(E)*(1+Z_1/Z_c)**-.5)

def Z_1(beta):
    return np.sin(beta/2)**2

def Z_c(E,F):
    return E**2/(4*abs(F))

def sn(u,m):
    return ellipj(u,m)[0]

def xi(tau,phi,m_xi,xi_minus):
    SN=sn(phi*tau,m_xi)
    return (xi_minus*SN**2)/((m_xi**-1)-SN**2)
```

```python
def cn(u,m):
    return ellipj(u,m)[1]    

def m_plus(E,F,Z_2,Z_c):
    return (abs(E)/F)*(np.sqrt(Z_2/Z_c))

def theta(E,Z_2,Z_c):
    return np.sqrt(abs(E))*(Z_2/Z_c)**.25

def m_eta(E,Z_2,Z_c):
    return 0.5*(1-np.sign(E)*(Z_2/Z_c)**-.5)

def Z_2(beta):
    return np.cos(beta/2)**2

def eta(tau,theta,m_eta,m_plus):
    CN=cn(theta*tau,m_eta)
    SN=sn(theta*tau,m_eta)
    return m_plus*((1-CN)/SN)**2
    
```

```python
def n_plus(E,F,Z_2,Z_c):
    return E/F*(1-np.sqrt(1-Z_2/Z_c))

def psi(E,Z_2,Z_c):
    return (np.sqrt(E)/2)*np.sqrt(1+np.sqrt(1-Z_2/Z_c))

def n_eta(Z_2,Z_c):
    return 2*(1+(1-Z_2/Z_c)**-0.5)**-1

def eta_g(tau,psi,n_eta,n_plus):
    SN=sn(psi*tau,n_eta)
    return n_plus*SN**2/(1-SN**2)
```

```python
import pdb

def trajectories(tau=np.linspace(0,1E-5,100),F=-1.55E5/5.142E11,E=24.9E-3/27.211,N=10,beta0=0,escape_fac=4000):
    
    beta_c=2*np.arccos(-E/(2*np.sqrt(abs(F))))
    if E>0:
        beta_c=sc.pi
    
    beta=np.linspace(beta0,min(beta_c-0.001,sc.pi),N)
    
    eta_traj=[]
    xi_traj=[]
    Zc=Z_c(E,F)
    
    for Z1,Z2 in zip(Z_1(beta),Z_2(beta)):
        
        if abs(E)<=2*np.sqrt(Z2*abs(F)):
            ETA=eta(tau,theta(E,Z2,Zc),m_eta(E,Z2,Zc),m_plus(E,F,Z2,Zc))
       
        if abs(E)>2*np.sqrt(Z2*abs(F)):
            ETA=eta_g(tau,psi(E,Z2,Zc),n_eta(Z2,Zc),n_plus(E,F,Z2,Zc))
        
        XI=xi(tau,phi(E,Z1,Zc),m_xi(E,Z1,Zc),xi_minus(E,F,Z1,Zc))
        
#         pdb.set_trace()
        
        Z=(XI-ETA)/2
        
        if np.any(abs(Z)>escape_fac):
            ETA[np.min(np.where(abs(Z) > escape_fac)):]=np.nan
            XI[np.min(np.where(abs(Z) > escape_fac)):]=np.nan
        
        eta_traj.append(ETA)
        xi_traj.append(XI)
            
            
    return xi_traj, eta_traj, beta

```

```python
nm=5.291E-2

def r_par(xi,eta):
    r=np.sign(xi)*np.sign(eta)*np.sqrt(xi*eta)
    dr=np.ediff1d(r)
    d2r=np.append(np.zeros(1),np.ediff1d(np.sign(dr)))
    d2r=np.append(d2r,np.zeros(1))
    cross=np.where((r<100)&(d2r>1))[0]

    flag=False
    if len(cross)>0:
        x_point_old=0
        j=0
        for x_point in cross:
            if flag==False:
                x_point_b=x_point
            if x_point==x_point_old+1 and flag==False:
                x_point_b=x_point
                j=1
                flag=True
                continue
            if x_point==x_point_old+1 and flag==True:
                j+=1
                flag=True
                continue
            else:
                r[x_point_b+int(j/2):]=-r[x_point_b+int(j/2):]
                flag=False
                
            x_point_old=x_point
    return r*nm
    
def z_par(xi,eta):
    return ((xi-eta)/2)*nm

def CS(x,y,F=5E5,Q=-1,r_0=0.1):
    r=np.sqrt(x**2+y**2)
    return F*x+Q/r
```

```python
def coulomb_stark_traj(sub,F=-1.3E5/5.142E11,E=-20.9E-3):
    EF=2*27.211*np.sqrt(abs(F))
    EH=(E-EF)/27.211
    print(E)
    
    T_xi,T_eta,beta=trajectories(tau=np.linspace(0,1000,1000),F=F,E=EH,N=20,beta0=0,escape_fac=6000)

    scale=5000

    x = np.linspace(-scale*0.25,scale, 500)
    y = np.linspace(-scale*.75,scale*.75, 500)

    X, Y = np.meshgrid(x, y)
    Z = CS(X, Y,F=F)+29.9E-3/27.211385

    vmin=-20E-3/27.211385
    vmax=40E-3/27.211385
    
    plt.subplot(sub)
    
    plt.imshow(Z,extent=[min(x)*nm,max(x)*nm,min(y)*nm,max(y)*nm],aspect='auto',vmin=vmin,vmax=vmax)
    plt.contour(X*nm,Y*nm,Z,np.linspace(vmin,vmax,15),colors='k',linewidths=0.3,linestyles='solid')

    for traj_xi,traj_eta,b in zip(T_xi,T_eta,beta):
        c=plt.cm.autumn(b/(sc.pi))
#         r_par(traj_xi,traj_eta)
#         plt.xlim([0,2000])
        plt.plot(z_par(traj_xi,traj_eta),r_par(traj_xi,traj_eta),color=c,linewidth=.75)
        plt.text(-50,150,'$E_{{exc}}$={0:2.2f} meV'.format(-E*1E3),color='w')

    plt.xlim([-scale*0.25*nm,scale*nm])
    plt.ylim([-scale*nm*.75,scale*nm*.75])
    plt.ylabel('x (nm)')
```

```python
E=np.linspace(3E-3,100E-3,3)
figs=[311,312,313]

for e,fig in zip(E,figs):
    coulomb_stark_traj(fig,E=e)
plt.xlabel('z (nm)')
print(E)
    
plt.show()
    
# plt.savefig('coulomb-stark.pdf',bbox_inches='tight')
```

```python
def m_sn(phi,m_eta,theta,m_xi):
    return ellipj(2*phi*ellipk(m_eta)/theta,m_xi)[0]

def vr(theta,m_plus,xi_minus,m_xi,m_sn):
    return theta*np.sqrt((1/m_plus)*(xi_minus*m_xi*m_sn**2)/(1-m_xi*m_sn**2))

def p_dist(E=5E-3,F=-1.3E5/5.142E11):

    EF=2*27.211*np.sqrt(abs(F))
    E=(E-EF)/27.211

    beta_c=2*np.arccos(-E/(2*np.sqrt(abs(F))))
    
    if E>0:
        beta_c=sc.pi
        
    beta=np.linspace(0,min(beta_c,sc.pi),50000)[:-1]

    Zc=Z_c(E,F)
    V_r=[]

    for Z1,Z2 in zip(Z_1(beta),Z_2(beta)):

        MSN=m_sn(phi(E,Z1,Zc),m_eta(E,Z2,Zc),theta(E,Z2,Zc),m_xi(E,Z1,Zc))

        V_r.append(vr(theta(E,Z2,Zc),m_plus(E,F,Z2,Zc),xi_minus(E,F,Z1,Zc),m_xi(E,Z1,Zc),MSN))
        
    
    return np.histogram(V_r,bins=500)

E=np.linspace(3E-3,15E-3,3)

for e in E:
    dist=p_dist(E=e)
    plt.plot(np.append(dist[1][1:],dist[1][-1]),np.append(dist[0],0)/50000,linewidth=0.5)
    text_x=dist[1][np.where(dist[0]==max(dist[0]))]
    text_y=max(dist[0])/50000
    plt.text(text_x+5E-4,text_y-1E-3,'{0:1.0f}'.format(e*1E3))

plt.xlabel('Transverse momentum (a.u)')
plt.ylabel('$P(\mathbf{p}_r)$')

plt.savefig('emission_angle_momentum.pdf',bbox_inches='tight')
```

```python
def vel_r(beta,E,F):
    Z1=Z_1(beta)
    Z2=Z_2(beta)
    Zc=Z_c(E,F)
    MSN=m_sn(phi(E,Z1,Zc),m_eta(E,Z2,Zc),theta(E,Z2,Zc),m_xi(E,Z1,Zc))
    return vr(theta(E,Z2,Zc),m_plus(E,F,Z2,Zc),xi_minus(E,F,Z1,Zc),m_xi(E,Z1,Zc),MSN)

def temp_Vr(E_range=np.linspace(1E-3,100E-3,30),F=-1.55E5/5.142E11,N=10):
    intergal=[]
    for E in E_range:
        EF=2*27.211*np.sqrt(abs(F))
        E=(E-EF)/27.211

        beta_c=2*np.arccos(-E/(2*np.sqrt(abs(F))))
        
        if E>0:
            beta_c=sc.pi
            
        beta=np.linspace(0,beta_c,N)[:-1]
        delta_beta=beta[1]-beta[0]
        
        I=1/(2*(1-np.cos(beta_c)))*np.trapz(vel_r(beta,E=E,F=F)**2*np.sin(beta),x=beta)
        
        
        intergal.append(I)
        
    return E_range,intergal,EF

EF_ran=[]


for F in np.linspace(-1E5,-10E5,4)/5.142E11:
    c=plt.cm.viridis(-(F/10E5)*5.142E11)
    Erange,intergal,EF=temp_Vr(E_range=np.linspace(1E-3,100E-3,60),N=5000,F=F)
    EF_ran.append(EF)
    plt.plot(Erange*1E3,np.array(intergal)*2.187691E6**2*sc.m_e/(sc.k),color=c,label='{0:1.1f} MV/m'.format(F*1E-6*5.142E11))
plt.plot(Erange*1E3,2/(3*sc.k)*Erange*sc.e,'--',color='C4',label='$2/3(k_B)E_{{exc}}$')
plt.plot(Erange*1E3,2/(3*sc.k)*(Erange-EF_ran[0])*sc.e,'-.',color='C1',label='$2/3(k_B)E_{{\lambda}}$')
plt.legend()
plt.ylim([0,800])
plt.xlabel('Excess energy (meV)')
plt.ylabel('Temprature (K)')

plt.savefig('temp_transverse_momentum.pdf',bbox_inches='tight')
```
